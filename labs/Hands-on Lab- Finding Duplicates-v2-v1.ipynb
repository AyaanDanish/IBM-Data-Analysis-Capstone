```python
# Load the dataset directly from the URL
file_path = "./survey_data_with_duplicate.csv"
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
print('Duplicated rows:', df[df.duplicated()].shape[0])
```

```python
df[df.duplicated()].head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
subset = ["MainBranch", "Employment", "RemoteWork"]

dupes = df.duplicated(subset=subset) # Get all the duplicate rows
duplicate_df = df[dupes]

# Groupby the duplicated rows, and check the unique count for all other columns
# Wherever the unique count is just 1, that means there's only one unique value there (meaning duplicates)
identical_counts = (duplicate_df.groupby(subset).nunique() == 1).sum().sort_values(ascending=False)
print(identical_counts)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
identical_counts.plot(kind='bar', figsize=(30, 10))
plt.title('Duplicate Count by Column', fontsize=20)
```

```python
duplicate_df.groupby('Country')['Country'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
duplicate_df.groupby('Employment')['Employment'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
# Define key columns for uniqueness
key_columns = ['ResponseId']

# Check for duplicates based on key columns
duplicates = df.duplicated(subset=key_columns)

# Print summary
print(f"Total rows: {len(df)}")
print(f"Duplicate rows based on key columns: {duplicates.sum()}")
print(f"Uniqueness percentage achieved: {((len(df) - duplicates.sum()) / len(df)) * 100:.2f}%")
```

## Verify and Document Duplicate Removal Process

2. Explain the reasoning behind selecting specific columns for identifying and removing duplicates.


- Using an ID column is the smartest move here because IDs are meant to be unique identifiers in a dataset.
- The ResponseId column had 20 duplicates, which might indicate that the same user made multiple responses to the survey sometimes.
- Removing these row will therefore remove all duplicate in the dataset
